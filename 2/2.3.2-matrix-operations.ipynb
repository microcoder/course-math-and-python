# Матричные операции

## Умножение матриц

В прошлый раз мы уже разобрались с вами, как умножать матрицу на вектор. А именно: мы можем умножать матрицу размера $ m \times n $ на вектор-столбец размера $ n \times 1 $. Для этого мы домножаем каждый элемент i-той строки на соответствующий элемент вектора и складываем. В итоге получаем i-тый элемент вектора-результата. Получаем мы на выходе вектор размера $ m \times 1 $:

* Умножение матрицы $ m \times n $ на вектор $ n \times 1 $

$$ Aw =
\begin{pmatrix}
\sum\limits_{i=1}^n a_{1i} w_i \\
\sum\limits_{i=1}^n a_{2i} w_i \\
\dots \\
\sum\limits_{i=1}^n a_{mi} w_i \\
\end{pmatrix}
$$

Обратите внимание, мы могли умножать матрицу на вектор только в том случае, если число столбцов в матрице совпадало с длиной вектора. Иначе операция была не определена.

Давайте теперь поговорим о том, как ввести более общую операцию — умножение матрицы на матрицу. Она будет возможна только в том случае, если число столбцов в первой матрице совпадает с числом строк во второй матрице. Иначе эта операция будет не определена, делать её будет невозможно.

Пусть есть две матрицы — $A$ и $B$:

* Матрица $A$ имеет размер $ m \times n $
* Матрица $B$ имеет размер $ n \times k $
* Результатом будет матрица $C$ размера $ m \times k $

То есть строк в ней будет столько же, сколько в первой матрице, а столбцов столько же, сколько во второй. Чтобы перемножить матрицы $A$ и $B$ и получить $C$, мы делаем следующее:

* $C = AB$
* $c_{ij} = \sum\limits_{p=1}^n \alpha_{ip} b_{pj}$

i-тый j-тый элемент матрицы $C$ получается поэлементным умножением i-той строки матрицы $A$ на j-тый столбец матрицы $B$, что затем складывается. Именно так и получается матрица $C$.

### Пример

Пусть есть две матрицы, первая имеет размер 3 × 2, вторая 2 × 3. Результатом будет матрица размера 3 × 3:

$$
\begin{pmatrix}
1 & 2 \\
0 & 1 \\
10 & 0 \\
\end{pmatrix}
*
\begin{pmatrix}
1 & 0 & 0 \\
0 & 0 & 2 \\
\end{pmatrix}
=
\begin{pmatrix}
 & & \\
 & & \\
 & & \\
\end{pmatrix}
$$

Давайте получим её элемент с индексами 1, 1. Для этого нам понадобится первая строка матрицы $A$ и первый столбец матрицы $B$. Перемножаем их. Получаем `1 * 1 + 2 * 0`. Значит, элемент с индексами 1, 1 матрицы $C$ будет равен 1:

$$
\begin{pmatrix}
\color{red} 1 & \color{red} 2 \\
0 & 1 \\
10 & 0 \\
\end{pmatrix}
*
\begin{pmatrix}
\color{red} 1 & 0 & 0 \\
\color{red} 0 & 0 & 2 \\
\end{pmatrix}
=
\begin{pmatrix}
 \color{red} 1 & & \\
   & & \\
   & & \\
\end{pmatrix}
$$

Посчитаем элемент с индексами 1, 2. В результате перемножения `1 * 0 + 2 * 0`  получаем 0:

$$
\begin{pmatrix}
\color{red} 1 & \color{red} 2 \\
0 & 1 \\
10 & 0 \\
\end{pmatrix}
*
\begin{pmatrix}
1 & \color{red} 0 & 0 \\
0 & \color{red} 0 & 2 \\
\end{pmatrix}
=
\begin{pmatrix}
 1 & \color{red} 0 & \\
   & & \\
   & & \\
\end{pmatrix}
$$

Посчитаем элемент с индексами 1, 3. В результате перемножения `1 * 0 + 2 * 2`  получаем 4:

$$
\begin{pmatrix}
\color{red} 1 & \color{red} 2 \\
0 & 1 \\
10 & 0 \\
\end{pmatrix}
*
\begin{pmatrix}
1 & 0 & \color{red} 0 \\
0 & 0 & \color{red} 2 \\
\end{pmatrix}
=
\begin{pmatrix}
 1 & 0 & \color{red} 4 \\
   & & \\
   & & \\
\end{pmatrix}
$$

Наконец, чтобы получить элемент с индексом 2, 1, перемножаем `0 * 1 + 1 * 0` и получаем 0:

$$
\begin{pmatrix}
1 & 2 \\
\color{red} 0 & \color{red} 1 \\
10 & 0 \\
\end{pmatrix}
*
\begin{pmatrix}
\color{red} 1 & 0 & 0 \\
\color{red} 0 & 0 & 2 \\
\end{pmatrix}
=
\begin{pmatrix}
 1 & 0 & 4 \\
 \color{red} 0 & & \\
   & & \\
\end{pmatrix}
$$

И так далее.

## Почему именно так?

У вас, наверное, уже возник вопрос: почему мы ввели операцию умножения матриц именно так? Эта формула не кажется интуитивной, она не приходит в голову в первую очередь. Но, оказывается, у неё есть довольно понятный глубокий смысл. Давайте попробуем его понять. Мы уже выясняли, что матрица, по сути, задаёт линейное преобразование, линейную функцию из одного векторного пространства в другое. То есть, по сути, она принимает на вход один вектор и возвращает другой, уже другой разрядности. Например, матрица $A$ может задавать преобразование из пространства $V$ в пространство $W$, а матрица $B$ — задавать преобразование из пространства $W$ в пространство $T$.

* $A$ - преобразование из $V$ в $W$
* $B$ - преобразование из $W$ в $T$

Поскольку это функции, можем определить их композицию, $B$ от $A$, то есть сначала к вектору из пространства $V$ применяем преобразование $A$, потом преобразование $B$, и в итоге вектор из пространства $V$ переходит в пространство $T$.

* $BA$ - преобразование из $V$ в $T$

При этом оказывается, если мы перемножим матрицы $B$ и $A$, то полученное произведение $BA$ будет как раз задавать преобразование из пространства $V$ в пространство $T$, ровно то преобразование, которое является композицией $B$ и $A$.

* $BA$ - задает матрицу преобразования-композиции $A$ и $B$

Это одна из главных причин, почему умножение вводится именно так.

Есть ещё одна причина. Представьте, что у нас есть некоторая система линейных уравнений:

* Система линейных уравнений: $Ax = b$

Например, это может быть задача поиска линейного классификатора для предсказания возникновения рака. В этом случае в строках `A` записаны пациенты, для каждого пациента известно, какие мутации имеют место в его геноме, а в `B` записан правильный ответ: возникнет рак через 5 лет или не возникнет. Тогда `x` — это веса, с которыми складываются индикаторы наличия мутации в геноме, и получается некоторый ответ.

* Что, если $x = Bz$?

При этом, на самом деле, `x` может тоже выражаться через другой вектор `z`, то есть может иметь место уравнение $x = Bz$. Откуда оно может быть? Оно может задавать некоторые дополнительные ограничения. Например, что две определённые мутации должны носить одинаковый вклад в принятие решения, то есть должны иметь одинаковый вес. Это всё можно отрегулировать матрицей $B$.

* Подставим: $A(Bz) = b$

В итоге получается, что нужно решать систему вида $A*(Bz) = b$. Оказывается, что если мы поставим выражение для вектора `x`, то есть $x = Bz$, в первое уравнение, раскроем скобки, сгруппируем и получим уравнение относительно `z`, то мы получим систему с матрицей, которая будет равняться произведению $A$ и $B$.

* Можем раскрыть скобки: $ABz = b$
* Теперь можем решать систему с матрицей: $AB$

То есть мы ввели умножение так, чтоб можно было подставлять одну систему линейных уравнений в другую, и при этом всё сохранялось.

## Сложение и умножение на число

Давайте поговорим о более простых операциях — сложение и умножение на число. Здесь всё гораздо более интуитивно. Чтобы сложить две матрицы, мы должны убедиться, что они имеют одинаковый размер.

* Можем складывать только матрицы одинакового размера

То есть обе имеют размерность $ m \times n $, иначе сложение не определено. Складываем мы поэлементно. То есть i-тый j-тый элемент матрицы $C$ суммы $A$ и $B$ будет равняться $ \alpha_{ij} + b_{ij}$. Всё очень просто.

* Сложение производится поэлементно:
  * $C = A + B$
  * $c_{ij} = \alpha_{ij} + b_{ij}$

То же самое касается умножения на число. Чтобы умножить матрицу `b` на некоторое число $\alpha$, просто умножаем на $\alpha$ каждый элемент. То есть $c_{ij}$ — результат — будет равен $\alpha * b_{ij}$

* Умножение производится поэлементно:
  * $C = \alpha B$
  * $c_{ij} = \alpha b_{ij}$

## Транспонирование

Наконец, поговорим о такой операции, как транспонирование. Мы будем с ней сталкиваться довольно часто. Для этого нам понадобится понятие главной диагонали. Пусть есть некая матрица $A$ размера 4 × 3:

$$
A = 
\begin{pmatrix}
\color{red} 1 & 4 & 1 \\
2 & \color{red} 3 & 3 \\
2 & 1 & \color{red} 0 \\
5 & 1 & 1 \\
\end{pmatrix}
$$

Главной диагональю называются все её элементы, у которых первый индекс равен второму, то есть все элементы вида $\alpha_{ii}$. В нашем случае, это будет элемент (1, 1), (2, 2) и (3, 3), которые имеют значения 1, 3 и 0. Это действительно диагональ.

* Главная диагональ - $\alpha_{ii}$

Транспонирование — это, по сути, поворот относительно главной диагонали. Матрица $A$ размера $ m \times n $ превращается в матрицу $A$ транспонированное размера $ n \times m $. Кстати, $A$ транспонированное обозначается как $A$ с верхним индексом $T$.

* Транспонирование — поворот относительно главной диагонали
* Матрица $A$ размера $ m \times n $ превращается в $A^T$ размера $ n \times m $

Иными словами, i-тый j-тый элемент транспонированной матрицы равен j-тому i-тому элементу исходной матрицы.

* $\alpha_{i\color{red}j}^T = \alpha_{\color{red}ji}$

Ещё можно интерпретировать это так, что строки исходной матрицы становятся столбцами транспонированной матрицы.

* Строки становятся столбцами

Вернёмся к нашему примеру. Матрица размера 4 × 3 становится матрицей 3 × 4, при этом главная диагональ 1, 3, 0 остаётся на своём месте, а всё остальное поворачивается относительно неё. То есть, например, столбец 1, 2, 2 ,5 (первый столбец) становится первой строкой 1, 2, 2, 5. И так далее.

$$
A = 
\begin{pmatrix}
\color{red}1 & \color{red}4 & \color{red}1 \\
\color{orange}2 & \color{orange}3 & \color{orange}3 \\
\color{green}2 & \color{green}1 & \color{green}0 \\
\color{blue}5 & \color{blue}1 & \color{blue}1 \\
\end{pmatrix}
\longrightarrow
A^T =
\begin{pmatrix}
\color{red}1 & \color{orange}2 & \color{green}2 & \color{blue}5 \\
\color{red}4 & \color{orange}3 & \color{green}1 & \color{blue}1 \\
\color{red}1 & \color{orange}3 & \color{green}0 & \color{blue}1 \\
\end{pmatrix}
$$

## Резюме

Что мы узнали?

* Умножение матриц соответствует композиции двух линейных преобразований
* Умножение на скаляр
* Траспонирование - перемена строк и столбцов местами

Умножение матриц — это довольно хитрая операция, которая при этом имеет довольно глубокий смысл. Например, она соответствует композиции двух линейных преобразований или соответствует постановке в систему линейных уравнений вектора, который тоже выражается как матрица умножить на вектор. При этом сложение и умножение для матриц (умножение на константу) определяется интуитивно, просто поэлементно. Также мы изучили с вами транспонирование — это операция, которая меняет строки и столбцы матрицы местами.

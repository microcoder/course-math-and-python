# Задача нахождения экстремума

**Экстремум** - это минимум (min) или максимум (max) у функции.

<img src="img/extremum.png" width="600">

Минимум функции достигается в точке `x` в том случае, если в некоторой окрестности точки `x` функция не принимает значение меньше, чем `f(x)`. Аналогично можно ввести понятие максимума функции.

## Локальный экстремум

Один из видов экстремумов — это локальный экстремум. Это тот случай, когда минимум не является наименьшим значением функции или максимум не является наибольшим значением функции:

<img src="img/local_extremum.png" width="400">

В точке `a` и в точке `b` у нас экстремумы. В точке `a` — максимум, а в точке `b` — минимум. Но при этом отчётливо видно, что есть точки, в которых функция достигает значения больше, чем в точке `a`, и меньше, чем в точке `b`. Поэтому точки `a` и `b` не являются точками глобального экстремума, а являются точками локального экстремума.

## Глобальный экстремум

Глобальный экстремум — это как раз тот случай, когда минимум будет наименьшим значением, а максимум будет наибольшим значением.

<img src="img/global_extremum.png" width="400">

## Экстремум и производная

Разберёмся, как понятие экстремума связано с производной. Если касательная графику функции в данной точке направлена под ненулевым углом к оси `x`, то понятно, что в этой точке экстремума не будет, потому что касательная неплохо приближает функцию. Это означает, что при сдвиге в одну сторону значение функции увеличится, при сдвиге в другую сторону значение функции уменьшится. Это никак не совмещается с понятием экстремума.

<img src="img/extremum_and_derivative.png" width="400">

Тогда мы можем потребовать, чтобы в точках, которые мы подозреваем на то, что это точки-экстремумы, касательная была параллельна оси `x`:

<img src="img/extremum_and_derivative2.png" width="400">

## Необходимое условие экстремума

И действительно, мы с вами придумали необходимое условие экстремума, то есть равенство нулю производной `f'(x)`:

<img src="img/extremum_and_derivative3.png" width="400">

При этом это условие не является достаточным, что иллюстрирует опять-таки график на слайде:

<img src="img/extremum_and_derivative4.png" width="400">

Если мы рассмотрим график `y = x³`, то в точке `x = 0` касательная будет направлена вдоль оси `x`. Но при этом совершенно ясно, что это не является ни точкой минимума, ни точкой максимума. Ведь если мы сдвинемся в сторону увеличения `x`, значение функции увеличится. Если мы сдвинемся в сторону уменьшения `x`, значение функции уменьшится.

## Когда производная не равна нулю

Также производная не обязана равняться нулю в каждой точке экстремума, потому что в некоторых точках производная может попросту не существовать:

<img src="img/extremum_and_derivative5.png" width="400">

## Итог

* Экстремумы — это минимумы и максимумы функции
* Экстремумы бывают локальными и глобальными
* Необходимое условие экстремума — это равенство нулю производной функции в случае, если она, конечно же, определена.

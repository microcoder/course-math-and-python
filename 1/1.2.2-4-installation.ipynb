# Варианты установки

## Установка дистрибутива Anaconda

### Установка Anaconda в Windows

* Перейдите на сайт https://www.anaconda.com/ в раздел [download](https://www.anaconda.com/download/#windows)
* После того как файл инсталлятора загружен, его необходимо запустить
* В инсталяторе выбрать следующие опции:
  * Нажимаем `I Agree`, чтобы принять лицензию
  * Оставляем по умолчанию опцию `Just me (recommended)`
  * Оставляем как есть `Destination Folder`
  * `Advanced option` - оставляем как есть
  * Нажимаем `Next` и начнется процесс установки
  * Нажимаем `Finish`
* Идем в меню `Пуск` и ищем `Anaconda` и попытаемся запустить Python в интерактивном режиме, для этого выберем ярлык `IPython`
* Для проверки, в открывшейся консоли напишем тестовый код:
  ````python
  print('Hello, World!')
  ````
* Проверим наличие необходимых нам библиотек. Это можно проверить с помощью команды `import`:
  ````python
  import pandas
  ````
  Далее посмотрим какая версия библиотеки установлена:
  ````python
  pandas.__version__
  ````
* Проверим, работает ли у нас Jupyter Notebook, для этого выбираем ярлык запуска `Jupyter`

### Установка Anaconda в Linux

* Перейдите на сайт https://www.anaconda.com/ в раздел [download](https://www.anaconda.com/download/#linux) и выбрать вариант загрузки
* Загрузить файл
  ````bash
  cd ~/Downloads
  wget -c https://repo.continuum.io/archive/Anaconda3-5.0.1-Linux-x86_64.sh
  ````
* После загрузки файла выполняем команду установки:
  ````bash
  bash Anaconda3-5.0.1-Linux-x86_64.sh
  ````
* Далее нам предложат ознакомится с лицензией, в конце ввыодим `yes`
* Затем предлагается указать путь по которому будет установлена Anaconda.
* Убедимся, что наш python запускается из Anaconda выполнив команду в консоли:
  ````bash
  which python3
  ````

## Установка вручную
### Установка Jupyter Notebook

* Установка Jupyter Notebook в системе Linux (Debian)
  ````bash
  sudo apt install python3-pip -y
  sudo pip3 install --upgrade pip
  sudo pip3 install jupyter
  ````
* Запуск Jupyter Notebook от имени простого пользователя системы:
  ````bash
  jupyter-notebook
  ````

### Установка модулей python

Установить следующие модули:

````bash
sudo pip3 install matplotlib
sudo pip3 install pandas
sudo pip3 install scipy
````

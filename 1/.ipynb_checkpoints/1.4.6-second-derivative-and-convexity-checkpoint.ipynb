# Вторая производная и выпуклость

Как влияет знак производной на рост функции?

* Если производная не отрицательна, то функция возрастает
* Если производная не положительна, то функция убывает, что понятно из направления касательной к графику функции в точке.

<img src="img/second_derivative.png" width="600">

## Выпуклость

Но что же будет, если сама производная будет возрастать или убывать? Какому свойству функции это будет соответствовать? Оказывается, что есть такое свойство. Это свойство выпуклости и свойство вогнутости:

<img src="img/second_derivative2.png" width="600">

Выпуклой функцией называется функция, для которой касательная постоянно увеличивает свой угловой коэффициент при увеличении `x`, а вогнутой функцией называется функция, для которой касательная будет уменьшать угловой коэффициент с ростом `x`. 

Также в некоторых старых источниках используется терминология «выпуклая вниз функция» и «выпуклая вверх функция». Но мы будем использовать современную терминологию.

## Выпуклость и вторая производная

Когда функция выпуклая, наклон касательной растет и производная растет. Следовательно, вторая производная должна быть не отрицательной. 

* Функция выпукла => наклон касательной (производная) растёт => вторая производная больше либо равна нулю:
  $$ f''(x) \geq 0 $$

Аналогично для вогнутой функции — вторая производная должна быть не положительной:

* Функция вогнута:
  $$ f''(x) \leq 0 $$
  
  
Если мы хотим потребовать строгой выпуклости, то есть отсутствия прямых участков на графике функции, то достаточно потребовать, чтобы производная от производной (ну то есть вторая производная) была больше нуля:

* Функция строго выпукла (без прямых на графике):
  $$ f''(x) > 0 $$

Аналогично для вогнутости — достаточно потребовать, чтобы вторая производная была меньше нуля:

* Функция строго вогнута:
  $$ f''(x) < 0 $$

## Достаточное условие экстремума

Что интересно, используя свойства выпуклости, мы можем сформулировать достаточное условие экстремума. Если выполнено необходимое условие (производная равна нулю), и при этом вторая производная больше нуля, то мы имеем дело с выпуклой функцией. Это означает, что в точке у нас точно будет минимум:

<img src="img/second_derivative3.png" width="600">

Аналогично, если вторая производная меньше нуля и выполнено необходимое условие экстремума, то мы имеем дело с вогнутой функцией, и это означает, что у нас точно будет максимум:

<img src="img/second_derivative4.png" width="600">

## Выпуклость и необходимое условие экстремума

* Вывод: если функция выпуклая или вогнутая, необходимого условия экстремума становится достаточно

## Более общее определение выпуклости

Можно ввести и более общее определение выпуклости функции. Для этого рассмотрим отрезок, соединяющий какие-то точки графика функции:

<img src="img/second_derivative5.png" width="600">

* Если функция проходит под этим отрезком, будем называть такую функцию **выпуклой**
* Если функция проходит над этим отрезком (и только над), будем называть ее **вогнутой**

Такое определение сработает и для функций, у которых нет производной в некоторых точках. Например, можно разобраться в том, что функция `y = |x|` — это выпуклая функция:

<img src="img/second_derivative6.png" width="600">

## Единственность экстремума

Выпуклость или вогнутость на некотором интервале приводит к тому, что наименьшее значение на нём может быть только одно:

* В случае минимума функция должна быть выпукла в окрестности минимума
* В случае максимума функция должна быть вогнута в окрестности максимума

<img src="img/second_derivative7.png" width="600">

Иметь более одного максимума (хотя бы два), не имея никакого минимума непрерывная функция не может:

<img src="img/second_derivative8.png" width="600">

А разрыв, при котором будет два минимума - приведет к нарушению вогнутости/выпуклости функции. 

Может быть случай, когда есть два минимума, но при этом нет максимума. Это разрыв. Ну, например, у нас может быть некоторая вертикальная асимптота, и отдельные части графика будут выпуклыми. Но в случае разрыва у нас нарушится выпуклость всей функции.

## Итог

* Знак первой производной связан с монотонностью функции, то есть с возрастанием или убыванием
* Знак второй производной связан с выпуклостью функции
* Выпуклость и вогнутость может быть определена и в общем случае, когда не во всех точках есть производная функции
* Выпуклость и вогнутость обеспечивают единственность экстремума

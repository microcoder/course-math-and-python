# Что такое Python и почему мы выбрали именно его?

## Версия Python

Существует 3 версии языка:

* Python 1 (1994)
* Python 2 (2000)
* Python 3 (2008)

Последнии версии языка поддерживаются, однако между ними нет обратной совместимости.

## Лицензия

* Свободно распространяется
* Python Software Foundation license
  * Открытый исходный код
  * Разрешено использование языка для создания коммерческих продуктов


## Портируемость

* Python портирован и работает почти на всех известных платформах: Microsoft Windows, почти все UNIX(+ Linux, FreeBSD), Mac OS, Mac OS X

## Синтаксис

* Просто писать код
* Просто отлаживать код
* Просто читать код

## Режимы работы

* Стандартный режим

* Интерактивный режим

## Python Zen

* import this
* Красивое лучше, чем уродливое
* Простое лучше, чем сложное
* Читаемость имеет значение
* Если реализацию сложно объяснить, то идея плоха. Если реализацию легко объяснить, то идея, возможно, хороша

## Особые виды матриц: Линейные преобразования

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(12, 12))

#-------------------------------------------------pic 1
ax1 = fig.add_subplot(121, aspect='equal')
ax1.set_ylim(-4, 4)
ax1.set_xlim(-4, 4)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax1.spines['left'].set_position('zero')
ax1.spines['bottom'].set_position('zero')
# Eliminate upper and right axes
ax1.spines['top'].set_color('none')
ax1.spines['right'].set_color('none')

r1 = np.array([[0, 0],
               [0.5, 0],
               [0.5, 0.5],
               [0, 0.5]])
r2 = np.array([[0.5, 0],
               [1, 0],
               [1, 0.5],
               [0.5, 0.5]])
r3 = np.array([[0, 0.5],
               [0.5, 0.5],
               [0.5, 1],
               [0, 1]])
r4 = np.array([[0.5, 0.5],
               [1, 0.5],
               [1, 1],
               [0.5, 1]])

ax1.add_patch(mpl.patches.Polygon(r1, color='red'))
ax1.add_patch(mpl.patches.Polygon(r2, color='limegreen'))
ax1.add_patch(mpl.patches.Polygon(r3, color='blue'))
ax1.add_patch(mpl.patches.Polygon(r4, color='yellow'))


#-------------------------------------------------pic 2
ax2 = fig.add_subplot(122, aspect='equal')
ax2.set_ylim(-4, 4)
ax2.set_xlim(-4, 4)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax2.spines['left'].set_position('zero')
ax2.spines['bottom'].set_position('zero')
# Eliminate upper and right axes
ax2.spines['top'].set_color('none')
ax2.spines['right'].set_color('none')


b = np.array([[-1, 1],
              [-1, 3]])

# Перемножение матриц через метод dot():
ax2.add_patch(mpl.patches.Polygon(r1.dot(b), color='red'))
ax2.add_patch(mpl.patches.Polygon(r2.dot(b), color='limegreen'))
ax2.add_patch(mpl.patches.Polygon(r3.dot(b), color='blue'))
ax2.add_patch(mpl.patches.Polygon(r4.dot(b), color='yellow'))


#-------------------------------------------------- Arrow
transFigure = fig.transFigure.inverted()

coord1 = transFigure.transform(ax1.transData.transform([3, 1]))
coord2 = transFigure.transform(ax2.transData.transform([-3, 1]))

# print(coord1, coord2)

arrow = mpl.patches.FancyArrowPatch(
    coord1, coord2, transform=fig.transFigure,  # Place arrow in figure coord system
    connectionstyle="arc3,rad=-0.3",
    arrowstyle='simple', alpha = 0.5,
    mutation_scale = 25.0
)

fig.patches.append(arrow)




plt.show()
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(12, 12))

#-------------------------------------------------pic 1
ax1 = fig.add_subplot(121, aspect='equal')
ax1.set_ylim(-4, 4)
ax1.set_xlim(-4, 4)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax1.spines['left'].set_position('zero')
ax1.spines['bottom'].set_position('zero')
# Eliminate upper and right axes
ax1.spines['top'].set_color('none')
ax1.spines['right'].set_color('none')

r1 = np.array([[0, 0],
               [0.5, 0],
               [0.5, 0.5],
               [0, 0.5]])
r2 = np.array([[0.5, 0],
               [1, 0],
               [1, 0.5],
               [0.5, 0.5]])
r3 = np.array([[0, 0.5],
               [0.5, 0.5],
               [0.5, 1],
               [0, 1]])
r4 = np.array([[0.5, 0.5],
               [1, 0.5],
               [1, 1],
               [0.5, 1]])

ax1.add_patch(mpl.patches.Polygon(r1, color='red'))
ax1.add_patch(mpl.patches.Polygon(r2, color='limegreen'))
ax1.add_patch(mpl.patches.Polygon(r3, color='blue'))
ax1.add_patch(mpl.patches.Polygon(r4, color='yellow'))


#-------------------------------------------------pic 2
ax2 = fig.add_subplot(122, aspect='equal')
ax2.set_ylim(-4, 4)
ax2.set_xlim(-4, 4)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax2.spines['left'].set_position('zero')
ax2.spines['bottom'].set_position('zero')
# Eliminate upper and right axes
ax2.spines['top'].set_color('none')
ax2.spines['right'].set_color('none')

# Ортогональная матрица.
b = np.array([[2, 1],
              [1.5, 2]])

# Перемножение матриц через метод dot():
ax2.add_patch(mpl.patches.Polygon(r1.dot(b), color='red'))
ax2.add_patch(mpl.patches.Polygon(r2.dot(b), color='limegreen'))
ax2.add_patch(mpl.patches.Polygon(r3.dot(b), color='blue'))
ax2.add_patch(mpl.patches.Polygon(r4.dot(b), color='yellow'))


#-------------------------------------------------- Arrow
transFigure = fig.transFigure.inverted()

coord1 = transFigure.transform(ax1.transData.transform([3, 1]))
coord2 = transFigure.transform(ax2.transData.transform([-3, 1]))

# print(coord1, coord2)

arrow = mpl.patches.FancyArrowPatch(
    coord1, coord2, transform=fig.transFigure,  # Place arrow in figure coord system
    connectionstyle="arc3,rad=-0.3",
    arrowstyle='simple', alpha = 0.5,
    mutation_scale = 25.0
)

fig.patches.append(arrow)




plt.show()
```

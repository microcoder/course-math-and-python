Данные про эти бутылки имеют двумерную структуру. Дело в том, что бутылок много, и каждая из них описывается набором чисел, или признаками, числовыми характеристиками этих бутылок. Такие двумерные структуры встречаются постоянно, поэтому нужно как-то их характеризовать. Они называются матрицами. Матрицы — это, по сути, таблица с числами. Они обычно обозначаются большими буквами, например $A$:

$$ A = 
\begin{pmatrix}
12 & 7 & 21 & 31 & 11 \\
45 & -2 & 14 & 27 & 19 \\
-3 & 15 & 36 & 71 & 26 \\
4 & -13 & 55 & 34 & 15 \\
\end{pmatrix}
$$

Элементы же матрицы $A$ (большое) обозначаются буквами $\alpha$ (малое) с двумя нижними индексами: первый обозначает номер строки, второй — номер столбца. Например, $ \alpha_{12} $ — это элемент матрицы $A$ в первой строке, втором столбце. В нашем примере это 7. Элемент $ \alpha_{31} $ — это элемент матрицы $A$ в третьей строке и первом столбце, у нас это −3:

* Два индекса: строка и столбец
  * $ \alpha_{12} = 7$
  * $ \alpha_{31} = -3$

Наша матрица имеет размер 4 на 5, то есть это четыре бутылки вина, каждая из которых характеризуется пятью признаками. Пространство всех таких матриц будем обозначать как красивую букву $\mathbb{R}$ с верхним индексом 4 умножить на 5:

* Пространство всех матриц 4 на 5: $ \mathbb{R}^{4 \times 5}$

## Системы уравнения

Зачем нужны матрицы? Например, они используются для работы с системами линейных уравнений. Представьте, что у нас есть четыре бутылки вина, и для каждой из них мы знаем, подлинная они или нет, сделана из правильного сорта винограда или из какого-то дешевого — это подделка. Можно закодировать это с помощью вектора. Например, если первая и четвертая бутылки подлинные, а вторая и третья — это подделки, то закодируем это вектором `y` с элементами 1, 0, 0 и 1, где 1 — это подлинность, 0 — это подделка:

* $ y = \begin{pmatrix} 1 & 0 & 0 & 1 \\ \end{pmatrix}$

Мы можем потребовать следующее: будем производить линейную классификацию, то есть будем складывать значение всех признаков с некоторыми весами, которые мы обозначаем как w₁, w₂, w₃, w₄ и w₅, и будем требовать, чтобы такая взвешенная сумма равнялась номеру класса, то есть 1 - если бутылка подлинная, 0 - если это подделка. Получаем следующую систему уравнений:

$$
\left\{
\begin{array}
12_{w1} & + & 7_{w2} & + & 21_{w3} & + & 31_{w4} & + & 11_{w5} & = & 1 \\
45_{w1} & + & -2_{w2} & + & 14_{w3} & + & 27_{w4} & + & 19_{w5} & = & 0 \\
-3_{w1} & + & 15_{w2} & + & 36_{w3} & + & 71_{w4} & + & 26_{w5} & = & 0 \\
4_{w1} & + & -13_{w2} & + & 55_{w3} & + & 34_{w4} & + & 15_{w5} & = & 1 \\
\end{array}
\right.
$$

У неё могут быть некоторые проблемы. Например, она может быть неразрешима или, наоборот, у нее может быть слишком много решений, но с этим будем разбираться позже, в курсах на машинное обучение. А пока будем рассматривать именно такую систему уравнений. Они будут встречаться постоянно в нашем курсе в самых разных задачах. И оказывается, можно очень удобно и лаконично записать такую систему, используя векторы и матрицы. При этом решения таких систем также будут выражаться через векторы и матрицы.

## Векторы и матрицы

Для начала разберемся с парой нюансов, а именно: давайте заметим, что вектор размера `n` — это, по сути, тоже матрица, у которой одна из размерностей равна 1.

* Вектор размера n - тоже матрица

При этом вектор может быть как вектор-столбцом, то есть матрицей размера `n` на 1, так и вектором-строкой, то есть матрицей размера 1 на `n`:

* Вектор-столбец: $ w = \begin{pmatrix} w_1 \\ w_2 \\ w_3 \\ w_4 \\ w_5 \\ \end{pmatrix}$
* Вектор-строка: $ w = \begin{pmatrix} w_1 & w_2 & w_3 & w_4 & w_5 \\ \end{pmatrix}$

## Умножение

Давайте введём операцию умножения матрицы размера `m` на `n` на вектор-столбец размера `n`:

* Умножение матрицы $ m \times n $ на вектор-столбец $ n \times 1 $:
$$ Aw =
\begin{pmatrix}
\sum\limits_{i=1}^n a_{1i} w_i \\
\sum\limits_{i=1}^n a_{2i} w_i \\
\dots \\
\sum\limits_{i=1}^n a_{mi} w_i \\
\end{pmatrix}
$$

Как это делается? Результатом умножения такой матрицы на такой вектор будет новый вектор длины `m`. При этом первый элемент этого вектора вычисляется как произведение каждого элемента первой строки матрицы $A$ на каждый элемент вектора `w`, и потом это суммируется. Второй элемент этого вектора — это произведение второй строки матрицы $A$ на вектор `w` и затем снова суммирование, и так далее.

Благодаря этому мы сможем матрично записать нашу систему линейных уравнений. Она будет выглядеть как $A$ умножить на `w` равняется `y` — всё очень просто и кратко.

* Матричная запись системы уравнений: $ Aw = y $

Давайте потренируемся умножать матрицы на векторы. Умножим нашу матрицу признаков $A$ на вектор весов с координатами 1, 2, 1, 0, 0:

$$
\begin{pmatrix}
\color{red}{12} & \color{blue}7 & \color{green}{21} & \color{orange}{31} & \color{fuchsia}{11} \\
45 & -2 & 14 & 27 & 19 \\
-3 & 15 & 36 & 71 & 26 \\
4 & -13 & 55 & 34 & 15 \\
\end{pmatrix}
\times
\begin{pmatrix}
\color{red}1 \\ \color{blue}2 \\ \color{green}1 \\ \color{orange}0 \\ \color{fuchsia}0 \\
\end{pmatrix}
=
\begin{pmatrix}
47 \\ 55 \\ 63 \\ 33 \\
\end{pmatrix}
$$

Найдем первый элемент их произведения. Он получается умножением первой строки матрицы на этот вектор, то есть 12 умножить на 1, плюс 7 умножить на 2, плюс 21 умножить на 1, плюс 31 умножить на 0, плюс 11 умножить на 0, и в сумме равно 47. Второй элемент вектора будет равен 55, третий — 63 и последний — 33.

```python
import numpy as np

a = np.array([[12, 7, 21, 31, 11],
              [45, -2, 14, 27, 19],
              [-3, 15, 36, 71, 26],
              [4, -13, 55, 34, 15]])
b = np.array([[1],
              [2],
              [1],
              [0],
              [0]])

print('shape a: ', a.shape, ', b: ', b.shape, '\n')

s = 0
for i in a:
    for k, j in enumerate(i):
        s += j * b[k][0]
        print('{} * {} = {}'.format(j, b[k][0], j * b[k][0]))
    print('Перемножение вектора = {}'.format(s))
    s = 0

# Тоже самое, но с помощью NumPy:
print('\nРезультат с помощью NumPy:\n', a1.dot(b1)) # dot() - перемножение матриц
```

## Линейное преобразование

Давайте заметим ещё один момент. Пусть есть некоторая матрица размера `m` на `n`. Умножая её на некоторый вектор длины `n`, мы получаем на выходе вектор длины `m`:

* Матрица $ m \times n $
* Умножаем на вектор длины $n$, получаем вектор длины $m$

В нашем случае из вектора длины 5 мы получили вектор длины 4. Получается, что матрица задаёт некоторое преобразование, некоторую функцию: из одного векторного пространства размерности `n` в другое векторное пространство размерности `m`.

* Матрица задает линейное преобразование

Они будут часто возникать у нас. Кстати, такие преобразования называются линейными.

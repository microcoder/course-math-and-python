# Знакомство с линейной алгеброй

## Векторы и матрицы

Все мы с вами помним из школы, что вектор — это стрелочка, которая соединяет две точки:

<img src="img/linalg1.png">

Возможно, вы знаете, что матрица — это таблица, состоящая из чисел:

$$
A=
\begin{Bmatrix}
    a_11 & a_12 & \dots & a_1n \\
    a_21 & a_22 & \dots & a_2n \\
    \vdots & \vdots & \ddots & \vdots \\
    a_n1 & a_n2 & \dots & a_nn \\
\end{Bmatrix}
$$

Но в каком виде они понадобятся нам? Какими свойствами они обладают? Почему они задаются именно так? С этим мы и разберемся в этом уроке.

## Признаковое описание

Представьте, что у нас есть сеть магазинов, и для каждого из них нужно предсказать прибыль в следующем месяце. Зачем это нужно? Например, зная, что в одном из магазинов прибыль упадет, мы можем принять меры и избежать этого.

В этой задаче магазины являются объектами. Это то, для чего мы делаем предсказание. Компьютер не понимает, что такое магазин. Ему нужно описать каждый объект с помощью чисел. Эти числа называются признаками, характеристиками объекта.

* Задача: предсказать прибыль магазина в следующем месяце
* Магазин - это объект
* Признаки - числовые характеристики объектов

Какие признаки могут быть в этой задаче? Давайте попробуем придумать:

По этому вектору будет делаться предсказание, заболеет он или нет.

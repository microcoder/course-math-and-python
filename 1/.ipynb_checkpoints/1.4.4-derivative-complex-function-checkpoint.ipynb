# Производная сложной функции

## Сложная функция

Сложная функция - это термин который используется для того, чтобы обозначить, что функция `f(x)` представима как функция `g`, взятая от значения другой функции `h(x)`:

$$ f(x) = g(h(x)) $$

Например, `ln(1 + x) = g(h(x))`, где `h(x) = 1 + x`, `g(h) = ln(h)`

## Дифференциал

Дифференциал вводится как линейная часть приращения функции, то есть фактически **производная функции** в точке `x * Δx` и его стоит рассматривать как функцию от `Δx`:

$$ df = f'(x) \Delta x $$

Но, если мы формально проделаем всё то же самое для функции `f(x) = x`, то мы увидим, что дифференциал этой функции будет в точности равен `Δx`:

$$ dx = x' \Delta x = \Delta x $$

Отсюда возникает мысль: вместо `Δx` в выражении для дифференциала писать сразу `dx`:

$$ df = f'(x)dx $$

И вот в такой форме, с выражением для дифференциала, мы обычно знакомимся и в таком виде используем.

## Производная и дифференциал

Легко заметить, что **производная** формально — просто частное `df / dx`. Таким образом, мы можем расписать **производную** сложной функции:

$$ f'(x) = \frac{df}{dx} $$

## Производная сложной функции

Нам нужно найти `dg` деленное на `dx`, и давайте просто умножим и поделим на `dh(x)`. В таком случае наше выражение просто распишется на произведение двух производных: `dg(h)` деленное на `dh` и `dh(x)` деленное на `dx`:

$$ \frac{dg(h(x))}{dx} = \frac{dg(h(x))}{dh(x)} \frac{dh(x)}{dx} = \frac{dg(h)}{dh} \frac{dh(x)}{dx} $$

Вот мы и получили простое правило, которое позволяет нам считать производную сложной функции. Более того, если наша функция еще более сложная, например `g(h(u(x)))`, это правило работает абсолютно так же. Нужно просто взять `dg` по `dh`, потом `dh` по `du`, потом `du` по `dx` и перемножить. То есть мы просто расписываем все по цепочке. Отсюда и англоязычное название `Chain Rule`.

## Итог 

Мы узнали, что такое **дифференциал** и почему **производная** часто записывается как `df / dx`. Кроме того, мы познакомились с `Chain Rule`, правилом взятия производной сложной функции, и это правило понадобится нам еще не раз. Ну, например, когда мы будем знакомиться с линейными классификаторами и c нейросетями.
